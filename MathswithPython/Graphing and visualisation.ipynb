```python
import matplotlib.pyplot as plt
import sympy as sym
import numpy as np
```

```python
x = sym.symbols('x')
y = x**2 - 3*x

for i in range(-11,12):
    plt.plot(i,y.subs(x,i),'o')

plt.xlabel('x')
plt.ylabel('f(x) = %s' %y)
plt.show()
```

```python
p1 = [-3,-1]
p2 = [4,4]

plt.plot([p1[0],p2[0]],[p1[1],p2[1]],color=[1,0,0],linewidth=2)
plt.axis('square')
plt.axis([-6,6,-6,6])
plt.show()
```

```python
# y = mx + b

x = [-5,5]
m = 2
b = 1

y =[340,2345]

for i in range(0,len(x)):
    y[i] = m*x[i] + b

plt.plot(x,y,label='y=%sx+%s' %(m,b))
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

# print(axis.get_xlim())

plt.legend()

plt.show()
```

```python
x = range(-3,4)
y = np.zeros(len(x))

for i in range(0,len(x)):
    y[i] = 2 - x[i]**2
    
plt.plot(x,y,'s-')
plt.show()
```

```python
x = np.linspace(-3,3,10)
y = 2 + np.sqrt(abs(x))

plt.plot(x,y,'ms-')
plt.show()
```

```python
x = np.linspace(-4,4,100)
e = range(-1,4)

for i in e:
    y = x**i
    plt.plot(x,y,label='$y=x^{%s}$' %i,linewidth=4)
    
plt.legend()
plt.xlim(x[0],x[-1])
plt.ylim(-20,20)
plt.show()
```

```python
from sympy.abc import x
import sympy.plotting.plot as symplot

y = x**2

p = symplot(y,show=False)
p.xlim=[0,50]
p[0].line_color = 'm'
p.title = 'This is a plot'

p.show()
```

```python
x,a = sym.symbols('x,a')

expr = a/x

n = 1

p = symplot(expr.subs(a,n),(x,-5,5),show=False)
p[0].label = '$y=%s$' %sym.latex(expr.subs(a,n))

p2 = symplot(expr.subs(a,2),show=False)
p.extend(p2)
p[1].label = '$y=%s$' %sym.latex(expr.subs(a,2))

p.ylim = [-5,5]
p.legend = True
p.show()

print(p)
```

```python
# from sympy.abc import a,x,y

y = a/(x**2 - a)


p = None
p = symplot(y.subs(a,1),(x,-5,5),show=False)
p[0].label = '$%s$' %sym.latex(y.subs(a,1))

for i in range(2,5):
    p.extend(symplot(y.subs(a,1),(x,-5,5),show=False))
    p[i-1].label = '$%s$' %sym.latex(y.subs(a,i))
    

p.ylim = [-10,10]
p.legend = True
p.show()
```

```python
A = np.zeros([10,14])

for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        A[i,j] = 3*i - 4*j

for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        plt.text(j,i,int(A[i,j]),horizontalalignment='center',verticalalignment='center')

# plt.set_cmap('Purples')
plt.imshow(A)
```

```python
N = 10

C = np.zeros((N,N))

for i in range(0,N):
    for j in range(0,N):
        C[i,j] = (-1)**(i+j)
        
plt.imshow(C)
plt.set_cmap('gray')
plt.tick_params(labelleft=False,labelbottom=False)
plt.show()
```

```python
from matplotlib.patches import Polygon
```

```python
y = np.array([ [1,1],[2,3],[3,1] ])
p = Polygon(y,facecolor='m',alpha=.3)

y1 = np.array([ [2,2],[2.5,4],[3.5,1] ])
p1 = Polygon(y1,facecolor='k',alpha=.3,edgecolor='k')

fig,ax = plt.subplots()

ax.add_patch(p)
ax.add_patch(p1)

ax.set_ylim([0,4])
ax.set_xlim([0,4])
plt.show()
```

```python
x = np.linspace(-2,2,101)
f = -x**2

pnts = np.vstack((x,f)).T
p = Polygon(pnts,facecolor='g',alpha=.2)

y = np.array([ [-.5,-4],[-.5,-2.5],[.5,-2.5],[.5,-4] ])
p1 = Polygon(y,facecolor='k')

fig,ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)

plt.plot(x,f,'k')
plt.plot([-2,2],[-4,-4],'k')
plt.axis('off')

# plt.savefig('figpic.png')

plt.show()
```
