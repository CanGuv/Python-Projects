```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
x = sym.symbols('x')

fx = x**3

lim_pnt = 1.5
lim = sym.limit(fx,x,lim_pnt)

display(Math('\\lim_{x\\to %g} %s = %g' %(lim_pnt,sym.latex(fx),lim)))
```

```python
fxx = sym.lambdify(x,fx)
xx = np.linspace(-5,5,200)

plt.plot(xx,fxx(xx))
plt.plot(lim_pnt,lim,'ro')

plt.show()
```

```python
fx = sym.sqrt(x+1)*sym.exp(-x)
gx = sym.cos(x+ sym.sin(x))

fxx = sym.lambdify(x,fx)
gxx = sym.lambdify(x,gx)

xx = np.linspace(0,10,100)

plt.plot(xx,fxx(xx),label='f(x)')
plt.plot(xx,gxx(xx),label='g(x)')

plt.legend()
plt.show()
```

```python
f = 3 + 2*x - 5*x**2 + 7*x**4
g = 4*x**2 + x**5

df = sym.diff(f)
dg = sym.diff(g)

d_fXg = sym.diff(f*g)
df_X_dg = sym.diff(f) * sym.diff(g)

d_fPlus_g = sym.diff(f+g)
```

```python
display(Math('\\text{Without applying the product rule}'))
display(Math('\\quad (f\\times g)\' = %s ' %sym.latex(sym.expand(d_fXg))))
display(Math('\\quad f\' \\times g\' = %s ' %sym.latex(sym.expand(df_X_dg))))
```

```python
display(Math('\\text{With applying the product rule}'))
display(Math('\\quad (f\\times g)\' = %s ' %sym.latex(sym.expand(d_fXg))))
display(Math('\\quad f\' \\times g + f\\times g\' = %s ' %sym.latex(sym.expand(df*g + f*dg))))
```

```python
display(Math('(f+g)\' = %s' %sym.latex(sym.expand(d_fPlus_g))))
display(Math('f\'+g\' = %s' %sym.latex(sym.expand(df + dg))))
```

```python
x = sym.symbols('x')

f = x**2
df = sym.diff(f)

xa = 1

fa = f.subs(x,xa)
dfa = df.subs(x,xa)

xx = np.linspace(-2,2,200)
f_fun = sym.lambdify(x,f)(xx)
df_fun = sym.lambdify(x,df)(xx)

tanline = dfa * (xx-xa) + fa

plt.plot(xx,f_fun,label='f(x)')
plt.plot(xx,tanline,label='tangent')
plt.plot(xa,fa,'ro')

plt.axis('square')
plt.axis([-2,2,-.5,2])

ax = plt.gca()
plt.plot(ax.get_xlim(),[0,0],'k--')
plt.plot([0,0],ax.get_ylim(),'k--')
plt.legend()
plt.show()
```

```python
from scipy.signal import find_peaks
```

```python
x = np.linspace(-5,5,1001)
fx = x**2 * np.exp(-x**2)
dfx = np.diff(fx)/(x[1]-x[0])

localmax = find_peaks(fx)[0]
localmin = find_peaks(-fx)[0]

print('The critical points are ' + str(x[localmax]) + ' ' + str(x[localmin]))

plt.plot(x,fx)
plt.plot(x[0:-1],dfx)

plt.plot(x[localmax],fx[localmax],'ro')
plt.plot(x[localmin],fx[localmin],'go')
plt.show()
```

```python
a,x = sym.symbols('a,x')

baseexpr = x**2 * sym.exp(-a*x**2)

arange = np.arange(0,2.25,.25)
xrange = np.linspace(-3,3,100)

fig,ax = plt.subplots(1,2)

for ai in arange:
    fx = baseexpr.subs(a,ai)
    dfx = sym.diff(fx)
    critpoints = sym.solve(dfx)
    
    ax[0].plot(xrange,sym.lambdify(x,fx)(xrange))
    ax[1].plot(xrange,sym.lambdify(x,dfx)(xrange))
    
    if 1 in critpoints:
        display(Math('\\Rightarrow %s \\text{ has a critical point at x=1}' %sym.latex(fx)))
    elif 2 in critpoints:
        display(Math('\\Rightarrow %s \\text{ has a critical point at x=2}' %sym.latex(fx)))
    else:
        display(Math('\\quad %s \\text{ has NO critical point at x=1 or x=2}' %sym.latex(fx)))
    
ax[0].set_ylim([0,2])
ax[0].set_title('Function')
ax[0].plot([1,1],[0,2],'--',color='gray')
ax[0].plot([2,2],[0,2],'--',color='gray')

ax[1].set_ylim([-2,2])
ax[1].set_title('Its derivative')
ax[1].plot([1,1],[-2,2],'--',color='gray')
ax[1].plot([2,2],[-2,2],'--',color='gray')

plt.show()
```

```python
from matplotlib.patches import Polygon

x = sym.symbols('x')

symf = x**2
symg = x

f = sym.lambdify(x,symf)
g = sym.lambdify(x,symg)

xx = np.linspace(-2,2,55)

xpatch = np.linspace(0,1,100)
ypatch = np.vstack((g(xpatch),f(xpatch))).T

fig,ax = plt.subplots()
ax.add_patch(Polygon(ypatch,facecolor='k',alpha=.3))

plt.plot(xx,f(xx))
plt.plot(xx,g(xx))

plt.legend(['$f(x)=%s$' %sym.latex(symf), '$g(x)=%s$' %sym.latex(symg)])
plt.axis([-.25,1.25,-.5,1.5])
plt.show()
```

### Computing the are between two functions
The area between two functions is givemn by the formula $A = \int_{a}^{b}[f(x)-g(x)]dx$

In our example, $f(x)=x^2$ and $g(x)=x$

Therefore, $A = \int_{a}^{b}(x-x^2)dx$

We will compute the are between the two crossing points, that is, where the two functions are equal. This is given by the two solutions to the equation
$$x^2=x$$

The two solutions are $x=0$ and $x=1$. This gives the definite integral of $$ = \int_{0}^{1}(x-x^2)dx$$

```python
fg_intersect = sym.solve(symf-symg)
display(Math('f(x)\\text{ and } g(x) \\text{ intersect at }x= %g\\text{ and }%g' %(fg_intersect[0],fg_intersect[1])))

A = sym.integrate(symg - symf,(x,fg_intersect[0],fg_intersect[1]))
display(Math('\\text{The area between these functions is }%g' %A))
```
