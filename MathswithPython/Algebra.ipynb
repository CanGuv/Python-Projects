```python
z = np.complex(4,2)
w = np.complex(3,5)

left_numer = (z-1) * (z*w+w)
left_denom = w*z - w
left_part = left_numer / left_denom

right_numer = (w-1) * (1-w)
right_denom = -w**2*z - w**2 + 2*w*z + 2*w - z - 1
right_part = right_numer / right_denom

solution = left_part + right_part

print(solution)
```

```python
x = np.linspace(0,2*np.pi,20)
fx = -(np.cos(x) + np.sqrt(x))

fmax = np.max(fx)
fmaxidx = np.argmax(fx)
print(fmax, fmaxidx)

plt.plot(x,fx,'bo-')
plt.plot(x[fmaxidx],fmax,'rs')

plt.show()
```

```python
from scipy.signal import find_peaks

x = np.linspace(0,12*np.pi,213)
fx = -np.cos(x) - np.sqrt(x)

peaks = find_peaks(fx)
print(peaks[0])

plt.plot(x,fx)
plt.plot(x[peaks[0]],fx[peaks[0]],'o')
plt.show()
```
